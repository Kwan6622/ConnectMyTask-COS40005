```python
import pandas as pd
import random

# Number of samples
num_samples = 1000
categories = ['Cleaning', 'Delivery', 'Plumbing', 'Moving Help']
urgency_levels = [0, 1]

data = []

for _ in range(num_samples):
    # 1. Generate task features
    cat = random.choice(categories)
    distance = round(random.uniform(1, 20), 1)  
    duration = round(random.uniform(0.5, 5), 1) 
    urgent = random.choice(urgency_levels)
    
    # 2. Base logic
    if cat == 'Cleaning': base_price = 50
    elif cat == 'Delivery': base_price = 30
    elif cat == 'Plumbing': base_price = 80
    elif cat == 'Moving Help': base_price = 100
    
    calculated_price = base_price + (distance * 2) + (duration * 10)
    if urgent == 1:
        calculated_price += 20
        
    market_variance = random.uniform(0.80, 1.30) 
    
    # 3. Random hidden fees or discounts (traffic, tips, feeling generous)
    hidden_fees = random.randint(-15, 30)
    
    # 4. Final messy price
    final_price = round((calculated_price * market_variance) + hidden_fees, 2)
    
    # Ensure price doesn't go absurdly low or high just in case
    final_price = max(final_price, 15.00)
    
    data.append([cat, distance, duration, urgent, final_price])

# Create and save the new DataFrame
df = pd.DataFrame(data, columns=['Category', 'Distance_km', 'Duration_hours', 'Is_Urgent', 'Price'])
df.to_csv('task_pricing_data_randomized.csv', index=False)

print("New randomized dataset created: task_pricing_data_randomized.csv")
print(df.head())
```
